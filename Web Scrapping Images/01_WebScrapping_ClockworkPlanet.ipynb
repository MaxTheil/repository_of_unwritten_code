### Clockwork Planet

```python
from bs4 import BeautifulSoup

import requests
import urllib.request
import shutil
```

```python
url = 'https://mangakakalot.com/chapter/clockwork_planet/chapter_36'
response = requests.get(url)

html = BeautifulSoup(response.text, "html.parser")
```

```python
html
```

```python
pics = html.find_all('img')
```

```python
pics
```

```python
images = []

for x in pics:
    images.append(x['src'])
```

```python
image_info = [x for x in images if 's3.mkklcdnv3' in x]
```

```python
for i in range(0, len(image_info)):
    output = requests.get(image_info[i], stream=True)
    realname = 'CP_Chapter36_P'+str(i+1)
    
    file = open('{}.jpg'.format(realname), 'wb')
    
    output.raw.decode_content = True
    shutil.copyfileobj(output.raw, file)
    del response
```
